```python
class Zapateria:
    def __init__(self, nombre, marcasDisponibles, numeroTallas, ubicacion):
        self.nombre = nombre
        self.marcasDisponibles = marcasDisponibles
        self.numeroTallas = numeroTallas
        self.ubicacion = ubicacion

    def medirPie(self):
        print(f"Sugerencia: Basado en la medida, tu talla ideal es {self.numeroTallas[0]}.")

    def recomendarModelo(self, marca):
        if marca in self.marcasDisponibles:
            print(f"Recomendación: El modelo más vendido de {marca} está disponible en {self.nombre}.")
        else:
            print(f"Lo sentimos, la marca {marca} no está disponible actualmente.")

    def realizarCambio(self):
        print(f"El cambio de calzado se ha realizado con éxito en la sucursal {self.ubicacion}.")

# Uso de la clase
zapateria1 = Zapateria("Calzados VIP", ["Nike", "Adidas"], [25, 26, 27], "Centro Histórico")
print(f"Zapatería: {zapateria1.nombre}, Ubicación: {zapateria1.ubicacion}")
zapateria1.medirPie()
zapateria1.recomendarModelo("Nike")
zapateria1.realizarCambio()

print("\n" + "—" * 30)

zapateria2 = Zapateria("Sport World", ["Puma", "Reebok"], [28, 29, 30], "Plaza Norte")
print(f"Zapatería: {zapateria2.nombre}, Ubicación: {zapateria2.ubicacion}")
zapateria2.medirPie()
zapateria2.recomendarModelo("Puma")
zapateria2.realizarCambio()

print("\ncreador de software Alexander Salazar NL = 49")
```
